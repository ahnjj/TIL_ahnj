#### 연습문제 1
- 타이타닉 호 승객에 대해서 다음의 값 추출

```python
# 선실별(class) 인원수, 사망/생존(alive)인원수 : 임의 정렬
titanic[['class', 'alive']].value_counts()
```

#### 연습문제 2
- 성별에 따른 생존자수와 사망자수의 정보를 갖고 있는 데이터프레임 생성  
    - 0은 생존 1은 사망

```python
sex_alive = pd.DataFrame(titanic[['sex', 'alive']].value_counts())
sex_alive
```

#### 연습문제 3
- 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive)
- 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율 출력  
    - bins=[1,18,35,50,65,99]
    - labels=['미성년자','청년','중년','장년','노년']

```python
alive_group = pd.DataFrame(titanic[['alive', 'age']].value_counts())

bins=[0,18,35,50,65,99]
labels=['미성년자','청년','중년','장년','노년']
age_ctg = pd.cut(titanic['age'], bins, labels=labels)

df = pd.DataFrame({
    '생존':titanic['alive'],
    '연령대':age_ctg
})

# 비율
pd.DataFrame(df[['생존','연령대']].value_counts(normalize=True))
```
