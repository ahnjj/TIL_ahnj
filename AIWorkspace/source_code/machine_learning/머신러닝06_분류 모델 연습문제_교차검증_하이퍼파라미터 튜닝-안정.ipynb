- 데이터 세트 분리 : 학습용 데이터 세트 70%, 테스트용(평가용) 데이터 세트 30%  
- 사용하는 ML 알고리즘 : 의사결정 트리 알고리즘   
- 모델 평가 : 예측 성능 평가 - 정확도  

### sklearn 패키지의  breast_cancer 데이터셋 사용

- label이 0 이면 malignant 양성, 1 이면 benign 음성

```python
import pandas as pd
```

### (1) 데이터 세트 준비 : 유방암 데이터 세트

```python
# 데이터 로드
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer
```

```python
cancer.data.size
cancer.target.size
```

```python
# cancer data와 label 분리
cancer_data = cancer.data
cancer_label = cancer.target
```

```python
cancer_df = pd.DataFrame(data=cancer_data, columns=cancer.feature_names)
cancer_df.head()
```

```python
cancer_df.info()
```

### (2) 데이터 세트 분리 : 학습 데이터 / 테스트 데이터 세트 
- 테스트 데이터 세트 : 30% 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, yTrain, yTest = train_test_split(cancer_data,
                                                cancer_label,
                                                test_size=0.3,
                                                random_state=11)
```

```python
Xtrain.size
yTrain.size
```

### (3) 모델 학습 : 학습 데이터 세트 기반으로 ML 알고리즘을 적용하여 모델 학습

```python
cancer_dt_clf = DecisionTreeClassifier(random_state=11)

cancer_dt_clf.fit(Xtrain, yTrain)
```

### (4) 예측 수행 : 학습된 ML 모델을 이용해서, 테스트 데이터의 분류 예측 

```python
y_pred = cancer_dt_clf.predict(Xtest)
y_pred
```

### (5) 평가 : 예측 정확도 평가  
    - 예측된 결과값과 테스트 데이터의 실제 결과값과 비교해서 ML 모델 성능 평가
    - 예측 정확도

```python
from sklearn.metrics import accuracy_score
print(f'예측 정확도 : {accuracy_score(yTest, y_pred)}')
```

### 교차 검증 수행

```python
from sklearn.model_selection import cross_val_score
import numpy as np


scores = cross_val_score(cancer_dt_clf, cancer_data, cancer_label, cv=5)
# scores

for count, accuracy in enumerate(scores):
    print('교차검증 {0} 정확도 : {1: .4f}'.format(count, accuracy))
print('평균 정확도 : {0.4f}', format(np.mean(scores)))
```

```python
# cross-validation 수행 안했을 때 정확도 :  0.9473684210526315
# cross-validation 수행 했을 때 평균 정확도 0.9156031672100605
```

### 하이퍼 파라미터 튜닝  수행

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth': [2, 3],
    'min_samples_leaf' : [1, 5],
    'min_samples_split' : [2, 3, 5]
}

grid_dclf = GridSearchCV(cancer_dt_clf, param_grid=parameters, scoring='accuracy', cv=5)
grid_dclf.fit(Xtrain, yTrain)

print('GridSearchCV 최적 하이퍼 파라미터 : ', grid_dclf.best_params_)
print('GridSearchCV 최고 정확도 : {0:.4f}'.format(grid_dclf.best_score_))

# best estimator : {'max_depth': 3, 'min_samples_leaf': 5, 'min_samples_split': 2}
best_dclf = grid_dclf.best_estimator_

y_pred = best_dclf.predict(Xtest)
accuracy = accuracy_score(yTest, y_pred)

print('하이퍼 파라미터 튜닝 끝낸 최종 정확도 : {0:.4f}'.format(accuracy))
 
```

```python
# 하이퍼 파라미터 튜닝 전 정확도 : 0.9156031672100605
# 하이퍼 파라미터 튜닝 끝낸 최종 정확도 : 0.9415204678362573
```

```python
grid_dclf.cv_results_['params']
```

```python
grid_dclf.cv_results_['mean_test_score']
```

```python
# 결과를 df로 만들어서 확인
result_df = pd.DataFrame(grid_dclf.cv_results_['params'])

# 열 추가
result_df['mean_test_score'] = grid_dclf.cv_results_['mean_test_score']
result_df.sort_values(by='mean_test_score', ascending=False)
```
