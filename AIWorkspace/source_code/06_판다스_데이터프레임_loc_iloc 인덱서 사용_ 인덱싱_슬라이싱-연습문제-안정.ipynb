### 데이터프레임 인덱싱/슬라이싱 연습문제

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
import pandas as pd
import numpy as np
```

```python
# (1) 25,50 범위의 숫자를 생성 후 5행 5열로 배치
df = pd.DataFrame(np.arange(25,50).reshape(5,5),
                 index=['a','b','c', 'd', 'e'],
                  columns = ['A', 'B', 'C', 'D', 'E'])
df
```

```python
# 위 df에 대하여 아래 내용 인덱싱 (인덱서 우선 사용)
```

```python
# (2) 열 라벨 인덱싱, A열 추출 - 시리즈
df['A']
df.loc[:,'A']
```

```python
# (3) 0행부터 2행까지 추출 - 슬라이싱 사용(행 인덱싱)
df['a':'c']
df[0:3]
df.iloc[0:3]
```

```python
# (4) 열행 인덱싱 - 원소값 a행 A열 - 열우선 인덱싱
df.loc['a','A']
```

25

```python
# (5) 1행의 3,4열(D,E열) : 인덱서 사용
df.loc[['b'],['D','E']]
```

```python
# (6) df로 반환(행,열 모두 슬라이싱처리) : 인덱서 사용
# a~c행, B~D열 출력
df.iloc[0:3, 1:4]
df.loc['a':'c','B':'D']
```

```python
# (7) 끝에서 두번째 행부터 마지막행까지 - df로 반환
df.iloc[-2:]
```

```python
# (8) a 행의  A부터 B열 - 시리즈로 반환 (인덱서 사용)
df.loc['a','A':'B']
```

```python
# (9) a 행의  A부터 B열 df로 반환 (인덱서 사용)
df.loc[['a'], 'A':'B']
```

```python
# (10) 행,열 슬라이싱 처리 - df 형태로 나타남 (인덱서 사용)
# d~e행 D~E열 출력
df.loc['d':'e','D':'E']
```

```python
# (11) A열의 a,b행 출력 : 슬라이싱
#시리즈 출력
df.loc[['a','b'],'A']
```

```python
# (12) A열의 a,b행 출력 : 슬라이싱
#df 출력
df.loc['a':'b',['A']]
df.loc[['a','b'],['A']]
```

```python
# (13) A열의 값이 30 이상인 행의 모든 값 출력
```

```python
df.loc[df.A>30]
```

```python
# (14) A열의 값이 30 이상인 행의 A~C열 출력
```

```python
df.loc[df.A >= 30, 'A':'C']
```
