#### seaborn 스타일링 함수

1. set_style() : 그림의 전반적인 모양 스타일링  
    - background color, grid, spine, tick 등 설정   
    - Built-in Themes (내장 테마) 활용  
    -  5가지 기본 제공 테마 : darkgrid, whitegrid, dark, white, ticks.   
        - 기본값은 darkgrid, 원하는대로 변경 가능  
2. despine() 함수
    - 축선 표시 여부 결정 (right=T/F,left=T/F,top=T/F,bottom=T/F)  
        - 각 속성값에 따라 표시 설정   
            - True  : 표시 안 함  
            - False : 표시  
3. set_context() 함수   
    - 프리젠테이션이나 보고서와 같은 다양한 매체에 활용할 수 있도록 스타일링  
    - 전체 스케일 조정 : paper, notebook, talk, poster(기본값 : notebook)  
    - 스케일로도 전반적인 사이즈를 조정가능하지만, 폰트 크기 조정 파라미터 별도 사용 가능  
    - sns.set_context() 함수의 font_scale 파마미터 사용  

```python
# 스타일 설정 : 배경 설정 
# set_style() : darkgrid, whitegrid, dark, white, ticks)
sns.set_style('dark')
plt.figure(figsize=(6,4))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
# 스타일 설정 : 축선 표시 despine()
# (right=T/F,left=T/F,top=T/F,bottom=T/F)
# True : 표시
# False : 표시 안 함
sns.set_style('whitegrid')
plt.figure(figsize=(6,4))
plt.plot(x,y1, x,y2, x,y3, x,y4)

sns.despine(left=True, bottom=True)
plt.show()
```

```python
# style 설정 : set_context()
# paper, notebook, talk, poster
sns.set_context('notebook', font_scale=0.5) # style 설정 : font_scale
plt.figure(figsize=(6,4))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

2. KDE Plot
- Kernel Density Estimation (커널 밀도 추정) 
- 하나 또는 두 개의 변수에 대한 분포를 나타내는 그래프
- histplot은 절대량이라면 kdeplot은 밀도 추정치를 시각화
    - 연속된 곡선의 그래프로 출력

```python
# 밀도 추정 (density estimation)
- 관측된 데이터들의 분포로부터 원래 변수의 (확률)분포 특성을 추정하는 것
- 예 : 특정 지역의 이동인구 수
  - 어떤 날은 10명
  - 어떤 날은 100명 
  - 몇 명이라고 단정짓기 어려움
  - 일정 기간(개월, 년) 동안 관측한 데이터들을 분석해서
    - 분포 특성을 찾아내는 전체 과정을 밀도추정이라고 봄    
    
```

### lmplot()
- 스캐터 그래프를 그릴때 회귀선을 기본으로 출력해주는 그래프 함수
- 옵션
    - x, y, data, palette  
    - height  : 그래프 크기
        - size 사용시 오류 발생 : height로 변경해서 사용  
    - hue  : 카테고리 데이터를 분류하여 그룹화하되 하나의 그래프에 두개의 값을 모두 표현
        - hue='sex' : 하나의 그래프에 male, female 표시
    - col  : 카테고리 데이터를 분류하여 그룹화하되 그룹의 개수만큼 그래프를 생성해서 표현
        - col='sex' : 2개의 그래프로 분리. male 그래프, female 그래프
    - fit_reg=False : 회귀선 생략 가능
    - ci = None : 신뢰구간 없음 (95:95% 신뢰구간)
    - scatter_kws = {'s':50, 'alpha':1} : 점의 크기 및 투명도

### 참고
회귀 분석
- 하나 또는 그 이상의 독립변수와 종속변수에 대한 영향을 추정할 수 있는 통계 기법  
- 두 변수의 상관관계를 기본으로 하여  
- 하나의 1차 선형식으로 두 변수의 관계를 일반화하는 분석 방법  

예:    
    - 독립변수가 종속변수에 영항을 미치는가?   
    - 독립변수가 1증가할 때 종속변수는 얼마나 증가할 것인가?  

회귀선에서 가장 유용한 값은 기울기  
- 기울기 : 종속변수에 독립변수가 미치는 영향을 나타낸다  
- 기울기 클수록 종속변수에 미치는 영향, 관계의 정도가 커진다는 의미  
- 기울기 음(-)이라면 변화의 방향이 반대로 독립변수가 증가할 수록 종속변수는 감소하는 관계  

```python
# lmplot() : 산점도와 함께 선형 회귀선 표시
# 선 주변의  그림자 영역 : 회귀 추정한  값의 신뢰 구간의 영역 
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=5)
plt.show()
```

```python
# 참고 : regplot() 사용 (유사)
# 여러 옵션 다름 : size, height, hue 사용하면 오류
sns.set_style('darkgrid')
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# 신뢰구간 영역 설정 : 일반적으로 95%로 설정
# ci 옵션 사용  
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=5, ci=50)
plt.show()
```

```python
# 회귀선 생략 가능 : git_reg=False
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=5, fit_reg=False)
plt.show()
```

```python
# 카테고리 변수 smoker가 포함된 경우의 회귀선
# hue='smoker' 설정
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, height=5,
          scatter_kws={'s':20, 'alpha':0.6}, palette="Set1")
plt.show()

# 점 크기 및 투명도 설정 가능
```

```python
# 범주형 변수 사용 시 겹치지 않고 별도 그래프로 표시 : col=범주형값
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, height=5,
          scatter_kws={'s':20, 'alpha':0.6}, palette="Set1",
          col='smoker')
plt.show()
```

```python
# 범주형 변수 사용 시 겹치지 않고 별도 그래프로 표시 : col=범주형값
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, height=5,
          scatter_kws={'s':20, 'alpha':0.6}, palette="Set1",
          col='day')
plt.show()
```

```python
# 카테고리 값이 많은 경우 아래로 배치 가능 : col_wrap 옵션
# col_wrap=2
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, height=5,
          scatter_kws={'s':20, 'alpha':0.6}, palette="Set1",
          col='day', col_wrap=2)
plt.show()
```

sns.heatmap()  
- annot = True :  숫자 표시 여부 <br>
- ax = ax :  히트맵을 그릴 격자 <br>
- linewidths = 0.4, : 선의 굵기 <br>
- linecolor = 'white' : 선의 색깔 <br>

#### 마커 및 popup 설정
    - 마커(특정 위치를 표시하는 표식)
    - popup(마커 클릭시에 나타나는 정보) : 환경에 따라 지원안되는 경우도 있음

1. 마커 생성
    - folium.Marker(): 위,경도 값 리스트, popup, icon 설정
    - 서클 마커 : CircleMarker() : 위,경도 값 리스트, popup, radius, color, fill_color
2. 마커 생성 후 지도에 부착
    - folium.Marker().add_to(지도객체변수)
    - folium.CircleMarker().add_to(지도객체변수)

```python
# 서울특별시청 
# 위도 : 37.566345
# 경도 : 126.977893
seoul_cityhall_map = folium.Map(location=[37.566345, 126.977893], zoom_start=17)
seoul_cityhall_map
```

```python
# 서울시청 지도에 마커 표시 
# popup 출력 : '서울 시정'

# 덕수궁에 서클마커 표시
# popup 출력 : '덕수궁'
# 덕수궁 : 37.5658049,  126.975146000007

seoul_cityhall_map = folium.Map(location=[37.566345, 126.977893], zoom_start=17)

# 서울시청에 마커 표시 및 팝업 출력
folium.Marker([37.566345, 126.977893], popup='시청').add_to(seoul_cityhall_map)

# 덕수궁에 서클마크 표시 및 팝업 출력
folium.CircleMarker([37.5658049, 126.975146000007], 
                    popup='덕수궁',
                    radius=50,
                    color='red',
                    fill_color='blue').add_to(seoul_cityhall_map)

seoul_cityhall_map
```
