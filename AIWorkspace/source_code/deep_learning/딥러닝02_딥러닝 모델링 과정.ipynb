### 1. 데이터 준비

### 2. 모델 설정 (구성/설계)
- 신경망 프레임 생성   
  - **model = Sequential()**   
- 신경망에 층 추가   
  - **model.add(Dense(1, input_dim=4, activation='sigmoid'))**  
    - Dense(units, input_dim, activation)  
    - Dense(출력값수, 입력값수, 활성화 함수)  
    - Dense 클래스는 뉴런의 입력과 출력을 연결해 주는 역할  
      - 입력 뉴런이 4, 출력 뉴런이 8이면 총 연결선은 4x8=32  
      - 각 연결선은 가중치 포함  
      - 가중치가 높을수록 입력값이 출력값에 미치는 영향이 커지고, 낮을수록 영향이 작아짐   
    - 활성화 함수 
      - linear : 입력과 동일한 값을 출력  
      - sigmoid : 시그모이드 함수로 이진분류에서 출력층에 주로 사용 (확률을 반환)
      - softmax : 소프트맥스 함수로 다중클래스 분류문제에서 출력층에 주로 사용 (확률을 반환)
      - relu: Rectified Linear Unit 함수를 사용하여 출력값을 0 이상의 값으로 변환.
              - 은닉층에서 주로 사용  
      - tanh : 하이퍼볼릭 탄젠트 함수 사용. 출력값을 -1과 1 사이의 값으로 변환. 
               - 은닉층에서 많이 사용 
      - 회귀에서는 출력층에 활성화 함수를 안쓴다! model.add(Dense(1))


### 3. 모델 컴파일
- **model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])**
- 앞에서 설정한 모델이 효과적으로 구현될 수 있도록  
- 여러 가지 환경을 설정하면서 실행 가능한 상태가 되도록 만들어주는 단계  
- 어떤 오차 함수를 사용할지 : loss  
- 어떤 최적화 방법을 사용할지 : optimizer  
- 수행 결과 평가는 어떤 방법을 사용할지 : metrics : accuracy  
  

### 4. 모델 학습 (모델 실행)
- **model.fit(X_train, Y_trian, epochs=100, batch_size=10)**  
- epoch : 에포크  
    - 학습 프로세스가 모든 샘플에 대해 한 번 실행하는 것  
    - epoch= 100 : 각 샘플이 처음부터 끝까지 100번 재사용될 때까지 반복한다는 의미  
- batch_size = 10 : 샘플을 한 번에 몇 개씩 처리할 것인지 설정  
    - 전체 ~개의 샘플을 10개씩 끊어서 집어 넣으라는 의미  
    - 배치 사이즈가 너무 크면 학습 속도가 느려지고,   
    - 너무 작으면 각 실행 값의 편차가 생겨서 전체 결과값이 불안정해질 수 있음  
    - 사용하는 메모리가 감당할 만큼의 배치 사이즈로 설정

### 딥러닝 모델링 예제

#### 1. 데이터 준비
-  AND 논리 data

```python
X_train = [[0,0],[0,1],[1,0],[1,1]]
y_train = [0,0,0,1]
X_test = [[0,1],[1,1],[1,0],[0,0],[1,1]]
y_test = [0,1,0,0,1]
```

#### 2. 모델 설정(구성/설계) 

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential # Sequential : 신경망 프레임 모듈
from keras.layers import Dense # Dense : 신경망층을 정의하는 모듈
```

```python
# 신경망 프레임 생성
and_model = Sequential()

# 층(layer)추가
# 입력값 수 : 2개
# 출력값 수 : 1개
# 활성화 함수 : activation='sigmoid' (binary 분류: 0 or 1)
and_model.add(Dense(units=1, input_dim=2, activation='sigmoid'))
```

#### 3. 모델 컴파일 

```python
and_model.compile(loss='binary_crossentropy', # loss : 손실함수 
                  optimizer='adam', 
                  metrics='accuracy') 

# binary_crossentropy (이항 교차 엔트로피) : 이진 분류 시 사용
# loss (손실 함수/오차함수)
```

#### 4. 모델 학습

```python
and_model.fit(x=X_train, y=y_train, epochs=100, batch_size=4)
# epochs: 모든 데이터의 학습을 몇 번 수행할 것인가(반복횟수)
# batch_size : 한번에 처리하는 데이터수
```

#### 5. 모델 평가 

```python
# 모델 평가
print('score : ', and_model.evaluate(X_test, y_test))
```

```python
# and_model의 가중치 확인
and_model.weights
# 입력 2개에 대한 가중치 
```

```python
# 가중치와 bias 확인
and_model.get_weights() # 가중치 , bias
weights, bias = and_model.get_weights()
```

### summary() 설명  
- **Layer(type)**
    - 레이어의 이름과 타입
    - 자동으로 레이어 이름 지정
    - 따로 지정해주고 싶을때는 Dense에 파라미터로 name='지정하고싶은 이름' 설정
    - dense 이름의 Dense 레이어 생성됨

- **Output Shape**  
    - (None, 1) :  None개의 행과 1개의 출력값 의미
    - 행이 None으로 지정되는 이유
      - 데이터 수는 계속해서 추가될 수 있기 때문에    
      - 딥러닝 모델에서는 주로 행은 지정하지 않고 shape을 맞추는 작업을 많이 수행  

- **Param**
    - 파라미터의 수
    - 즉 각 입력노드와 출력노드에 대해 연결된 간선의 수
    - 입력이 2개, 출력이 1개 이므로 2*1=2개의 간선이 존재해야 하는데 
    - 3개의 간선으로 표시  
    - 이유 : 입력에 Bias(b) 노드가 추가되기 때문
    - 따라서 입력값(2)+바이어스노드(1)=3 총 3개의 입력 노드가 있으므로 
    - 파라미터 수는 3*1=3개가 됨

### Dense 층 추가 : 2개 층 모델

```python
# 1. 데이터 준비
X_train = [[0,0],[0,1],[1,0],[1,1]]
y_train = [0,0,0,1]
X_test = [[0,1],[1,1],[1,0],[0,0],[1,1]]
y_test = [0,1,0,0,1]
```

```python
# 2. 모델 설정(구성/설계) 
model = Sequential()

# 첫 번째 층은 입력층을 설정해야하고
# 두 번째 층부터는 설정하지 않아도 됨
# - 이유 : 이전 층의 출력이 다음 충의 입력이 되기 때문

# 첫 번째 층 : 입력 2개, 출력 1개
# 두 번째 층 : 입력 1개 (첫번째 층의 출력값), 출력 1개
model.add(Dense(units=1, input_shape=(2,), activation='sigmoid'))
model.add(Dense(units=1, activation='sigmoid'))


# 3. 모델 컴파일 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. 모델 학습
model.fit(x=X_train, y=y_train, epochs=2000, batch_size=4)
```

```python
# 5. 모델 평가 
print('score : ', model.evaluate(X_test, y_test))
```

```python
model.get_weights()
```

```python
# model모델의 summary 확인
model.summary()

# 모든 층마다 bias가 1개 입력
# (입력 + bias) X output = param
# 첫번째 층 : 2 + 1 = 3 -> 3 X 1 = 3
# 두번째 층 : 1 + 1 = 2 -> 2 X 1 = 2
```

### 입력수, 출력수에 따른 파라미터 계산 

```python
model2 = Sequential()
model2.add(Dense(units=5, input_shape=(1,), activation='sigmoid'))  # 5 +
model2.add(Dense(units=3, activation='sigmoid'))
model2.add(Dense(units=1, activation='sigmoid'))  # 최종
model2.summary()


# 모든 층마다 bias가 1개 입력 (입력 + bias) X output = param
# 첫 번째 층 : 1 개 입력,  5개 출력 : (1+1) X 5 = 10
# 두 번째 층 : 5 개 입력,  3개 출력 : (5+1) X 3 = 18
# 세 번째 층 : 3 개 입력,  1개 출력 : (3+1) X 1 = 4
# ---------------------------------------------
#                        total params : 32
```

## 연습 문제
- OR 논리의 신경망 모델 생성
- AND 모델 기준으로 데이터 준비해서 모델링 수행


```python
# 1. 데이터 준비
X_train = [[0,0],[0,1],[1,0],[1,1]]
y_train = [0,0,0,1]
X_test = [[0,1],[1,1],[1,0],[0,0],[1,1]]
y_test = [0,1,0,0,1]

# 2. 모델 설정
# 신경망 프레임 생성
or_model = Sequential()

# 층(layer) 추가
or_model.add(Dense(units=1, input_shape=(2,), activation='sigmoid'))
or_model.add(Dense(units=1, activation='sigmoid'))

# 3. 모델 컴파일
or_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. 모델 학습
or_model.fit(x=X_train, y=y_train, epochs=2000, batch_size=4)

# 5. 모델 평가
print('score : ', or_model.evaluate(X_test, y_test))
```
