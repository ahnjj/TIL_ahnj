## 첫 번째 딥러닝 예제 : 폐암 수술 환자의 생존율 예측하기 실습

폐암 수술 환자의 생존율 예측
- 데이터 준비
(1) 딥러닝 모델 설정
(2) 모델 컴파일
(3) 모델 실행 (학습)

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# 준비된 수술 환자 데이터를 불러 오기
Data_set = np.loadtxt("../../data/ThoraricSurgery.csv", delimiter=",")
Data_set.shape
Data_set[:3]
```

### (1) 데이터 준비

```python
# (1) 데이터 준비
# - (1) 환자 기록 : 수치
# - (2) 생존 여부
```

```python
# 데이터 분리 : 학습/테스트용 데이터 세트로 분리
# train_test_split() : 0.3

from sklearn.model_selection import train_test_split
feature_data = Data_set[:, :-1]
label_data = Data_set[:, -1:]

X_train, X_test, y_train, y_test = train_test_split(feature_data,
                                                    label_data,
                                                    test_size=0.3, 
                                                    random_state=11)
```

```python
# 실행할 때 마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)

# (2) 딥러닝 모델 설정
# 입력 수 : 17
# 출력 수 : 1
model = Sequential()
model.add(Dense(units=1, input_dim=17, activation='sigmoid'))

# (3) 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# (4) 모델 학습 (실행)
# epochs = 100
# batch_size = 10
model.fit(x=X_train, y=y_train, epochs=100, batch_size=10)
```

```python
# (5) 모델 평가
print('score : ', model.evaluate(X_test, y_test))
```

```python
# (6) summary 확인
model.summary()
```

### 2개 층으로 구성된 예측 모델

```python
# 첫 번째 층 : 입력 7개, 출력 30개
# 두 번째 층 : 출력 1개

# 입력을 7로 했더니 오류남, 17개로 수정해서 재 실행

model2 = Sequential()
model2.add(Dense(units=30, input_shape=(17,), activation='sigmoid'))
model2.add(Dense(units=1, activation='sigmoid'))

# (3) 모델 컴파일
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# (4) 모델 실행
model2.fit(x=X_train, y=y_train, epochs=100, batch_size=10)

# (5) 모델 평가
print('score : ', model2.evaluate(X_test, y_test))
```
