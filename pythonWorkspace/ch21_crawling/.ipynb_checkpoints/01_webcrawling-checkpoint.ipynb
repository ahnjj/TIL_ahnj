### urllib 패키지 사용한 소스 추출

```python
from urllib.request import urlopen
```

```python
# 응답 객체에서 소스 코드 가져오기
url = "http://www.tistory.com"
html = urlopen(url)
text = html.read()
```

```python
text.decode('utf8')
```

### BeautifulSoup 사용법

```python
import bs4
```

```python
url = "https://www.tistory.com"
html = urlopen(url)
```

```python
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
```

```python
#bs_obj  # tag형식으로 받아옴
```

```python
# 계층 구조로 표현 : 가독성이 좋음
print(bs_obj.prettify())
```
